```python
import sys
sys.path.append("../code")

from data import FloatingSeaObjectDataset
from visualization import plot_batch, calculate_fdi, ndvi_transform, s2_to_RGB
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N_PIXELS_FOR_EACH_CLASS_FROM_IMAGE = 5
```

```python
data_path = "/ssd/floatingObjects/data"
from visualization import calculate_fdi, ndvi_transform 
```

```python
image_size = (128, 128)

def sample_N_random(data, N):
    idxs = np.random.choice(np.arange(len(data)), min(len(data),N), replace=False)
    return data[idxs]

def aggregate_images(x,y):
    """
    aggregates images to pixel datasets
    x: D x H x W -> N x D where N randomly sampled pixels within image
    y: H x W -> N
    """
    N=N_PIXELS_FOR_EACH_CLASS_FROM_IMAGE

    floating_objects = x[:,y.astype(bool)].T
    not_floating_objects = x[:,~y.astype(bool)].T

    # use less if len(floating_objects) < N
    N = min(N, len(floating_objects))

    x_floating_objects = sample_N_random(floating_objects, N)
    y_floating_objects = np.ones(x_floating_objects.shape[0], dtype=int)

    x_not_floating_objects = sample_N_random(not_floating_objects, N)
    y_not_floating_objects = np.zeros(x_not_floating_objects.shape[0], dtype=int)

    x = np.vstack([x_floating_objects, x_not_floating_objects])
    y = np.vstack([y_floating_objects, y_not_floating_objects]).reshape(-1)
    return x,y

def feature_extraction_transform(x,y):
    x,y = aggregate_images(x,y)
    
    # transforms require band dimension first:
    # so N x D -transpose> D x N -ndvi/fdi> N
    ndvi = ndvi_transform(x.T)
    fdi = calculate_fdi(x.T)
    
    # N x 2
    return np.vstack([fdi, ndvi]).T, y

def s2_to_ndvifdi(x):
    ndvi = ndvi_transform(x)
    fdi = calculate_fdi(x)
    return np.stack([fdi, ndvi])


dataset = FloatingSeaObjectDataset(data_path, fold="train", transform=feature_extraction_transform, 
                                   output_size=image_size, hard_negative_mining=False, use_l2a_probability=1)
test_dataset = FloatingSeaObjectDataset(data_path, fold="val", transform=feature_extraction_transform, 
                                         output_size=image_size, hard_negative_mining=False, 
                                         use_l2a_probability=1)


imagedataset = FloatingSeaObjectDataset(data_path, fold="train", transform=None, 
                                   output_size=image_size, hard_negative_mining=False, use_l2a_probability=1)
```

```python
def draw_N_datapoints(dataset, N):
    idxs = np.random.randint(len(dataset), size=N)

    x = []
    y = []
    for idx in idxs:
        x_,y_,fid_ = dataset[idx]
        x.append(x_)
        y.append(y_)

    return np.vstack(x), np.hstack(y)
```

## Data Visualization

```python
x,y = draw_N_datapoints(dataset, N=200)
```

```python
floating_objects = x[y.astype(bool)]
not_floating_objects = x[~y.astype(bool)]

fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(floating_objects[:,0], floating_objects[:,1], alpha=0.5, s=2, color="#e37222")
ax.scatter(not_floating_objects[:,0], not_floating_objects[:,1], alpha=0.5, s=2, color="#0065bd")
ax.set_xlabel("FDI")
ax.set_ylabel("NDVI")
ax.legend(["floating object", "water"])
ax.set_ylim(-.6,.6)
ax.set_xlim(-.01,.15)
sns.despine()
```

## Support Vector Classification

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

### Train/val data

from same region

```python
x,y = draw_N_datapoints(dataset, N=400)

clf = svm.SVC(gamma=1000, C=10)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred, target_names=["floating objects","water"]))
```

```python
x,y,fid = imagedataset[0]
features = s2_to_ndvifdi(x)
y_pred = clf.predict(features.reshape(2,-1).T).reshape(128,128)

fig, axs = plt.subplots(1,5, figsize=(5*3, 3))

images = [s2_to_RGB(x), features[0], features[1], y, y_pred]
titles = ["RGB", "ndvi", "fdi", "labels", "SVM prediction"]
for ax, img, title in zip(axs, images, titles):
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(title)
```

### Test data

from different regions

```python
X_test,y_test = draw_N_datapoints(test_dataset, N=200)

y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred, target_names=["floating objects","water"]))
```

## Test
